```python
import random
import math
import names
import csv
import pandas as pd
import json
from datetime import date, timedelta
random.seed(69)

# add read and hover attributes to purchase data, remove quantity and shipping
with open(r'Purchase.csv') as f:
    line = list(csv.reader(f))
newdata=[]
for rows in line[1:]:
    del rows[2]
    del rows[-1]
    new_attribute=['hover',random.randint(10,100),'read',random.randint(10,100)]
    rows.extend(new_attribute)
    newdata.append(rows)
```

```python
# Generate hover but not read data
# Based on Dhruv's code
sdate = date(2000, 1, 1)   # start date
edate = date(2016, 12, 31)   # end date

delta = edate - sdate# as timedelta

date_list = []

for i in range(delta.days + 1):
    day = sdate + timedelta(days=i)
    dates = date_list.append(day)
#print(date_list)


## 30 transactions a day, 200 customers a year with 50 sales a year
## 4: Creating list of customer IDs

customers = []
for i in range(1, 201):
    name = names.get_full_name()
    name = name.split()
    customer_id = name[1] + name[0][0]
    customer_id = customer_id.lower()
    customers.append(customer_id)
#print(customers)

## 5: Creating Header
header = ['CustomerID', 'ProductID', 'Date', 'Shipping']
product_ids = list(range(1, 33))

## 6: Creating 2D data
data = []
#for i in range(10):
for i in range(random.randint(1000,10000*16)):
    productid = random.choice(product_ids)
    transaction = [random.choice(customers), productid,random.choice(date_list),
                   'hover',random.randint(10,100),'read',0]
    data.append(transaction)

## 7: Sorting by date
data = sorted(data, key=lambda x: x[3])

for row in data:
    row[2] = str(row[2].strftime('%d/%m/%Y'))

# adding these data to new data
for rows in data:
    newdata.append(rows)
```

```python
# Extract unique dates
unique_dates = set()
for row in newdata:
    unique_dates.add(row[2])
unique_dates = sorted(unique_dates)

# Generate hover and read times 
dates = []
for date in unique_dates:
    events = []
    dates.append({"date": date, "events": events})
    for row in newdata:
        if row[2] == date:
            events.append({"id": row[0], row[3]: row[1], "duration": row[4]})
            events.append({"id": row[0], row[5]: row[1], "duration": row[6]})
output = {"all": dates}
#print(output) # Check output

with open(r'web.json',"w") as f:
    json.dump(output,f,indent = 4)
```
